```python
def absolute_trick(y_intercept, slope, x, y, learning_rate):
    y_hat = slope*x + y_intercept
    if y > y_hat:
        slope += x * learning_rate
        y_intercept += learning_rate
    if y < y_hat:
        slope -= x * learning_rate
        y_intercept -= learning_rate
    return y_intercept, slope
```

```python
absolute_trick(3, 2, 5, 15, 0.01)
```

```python
def square_trick(y_intercept, slope, x, y, learning_rate):
    y_hat = slope*x + y_intercept
    slope += x * learning_rate * (y - y_hat)
    y_intercept += learning_rate * (y - y_hat)
    return y_intercept, slope
```
